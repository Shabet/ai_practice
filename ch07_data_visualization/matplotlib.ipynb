# 기본 그래프 그리기

```python
from matplotlib import pyplot as plt

plt.plot([8,9,10],[80,90,100])


plt.show()
```

# 타이틀과 라벨 설정하기

```python
from matplotlib import pyplot as plt

plt.plot([8,9,10],[80,90,100])
plt.xlabel('study time')
plt.ylabel('score')
plt.title('Correlation between study time and test score')
plt.show()
```

## 여러 데이터 표현하기



```python
from matplotlib import pyplot as plt

plt.plot([8,9,10],[80,90,100])
plt.plot([8,9,10],[40,45,50])
plt.xlabel('study time')
plt.ylabel('score')
plt.title('Correlation between study time and test score')
plt.legend(['mid-term','finals'])
plt.show()
```

## bar 차트 그리기

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(3)
years = ['S', 'A', 'B']
values = [100, 200, 300]

plt.bar(x, values)
plt.xticks(x, years)

plt.show()
```

## pie 차트 그리기

```python
import matplotlib.pyplot as plt

ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False)
plt.show()
```

## Histogram 그리기

```python
import matplotlib.pyplot as plt

weight = [68, 81, 64, 56, 78, 74, 61, 77, 66, 68, 59, 71,
          80, 59, 67, 81, 69, 73, 69, 74, 70, 65]

plt.hist(weight)

plt.show()
```

## Scatter 그리기

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)

n = 50
x = np.random.rand(n)
y = np.random.rand(n)


print(type(x))
plt.scatter(x, y)
plt.show()
```
